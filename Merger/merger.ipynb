```python
import pandas as pd
columns=['Name', 'Features', 'Location', 'Price', "Seller", "Auction_Type", "Date"] 

d1 = pd.read_csv("data1.csv", names=columns, header=None)
d2 = pd.read_csv("data2.csv", names=columns, header=None)
d3 = pd.read_csv("data3.csv", names=columns, header=None)
```

```python
frames = [d1, d2, d3]
result = pd.concat(frames)
```

```python
result = result.drop_duplicates()
result.to_csv("Porsche.csv",index=False)
```
